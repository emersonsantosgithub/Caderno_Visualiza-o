## Analisar as tendências de imigração do Brasil de imigração do Brasil em um determindado período ##

```python
# Exportando o arquivo em transformando em DataFrame

import pandas as pd 

df = pd.read_csv('canadian_immegration_data.csv')
```

```python
# Visualizando as 5 primeiras linhas do DataFrame
df.head(5)
```

```python
df.set_index('Country', inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brazil', anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(), 
               'imigrantes':brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[8,4])
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.title("Imigração do Brasil para o canáda")
plt.xlabel("ano")
plt.ylabel("Número de imigrantes")
plt.xticks(['1980', '1985', '1990','1995', '2000', '2005', '2010','2015'])
plt.show()
```
