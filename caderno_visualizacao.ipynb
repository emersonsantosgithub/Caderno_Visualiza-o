## Analisar as tendências de imigração do Brasil de imigração do Brasil em um determindado período ##

```python
# Exportando o arquivo em transformando em DataFrame

import pandas as pd 

df = pd.read_csv('canadian_immegration_data.csv')
```

```python
# Visualizando as 5 primeiras linhas do DataFrame
df.head(5)
```

```python
df.set_index('Country', inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
```

```python
brasil = df.loc['Brazil', anos]
```

```python
brasil_dict = {'ano':brasil.index.tolist(), 
               'imigrantes':brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=[8,4])
plt.plot(dados_brasil['ano'],dados_brasil['imigrantes'])
plt.title("Imigração do Brasil para o canáda")
plt.xlabel("ano")
plt.ylabel("Número de imigrantes")
plt.xticks(['1980', '1985', '1990','1995', '2000', '2005', '2010','2015'])
plt.show()
```

```python
dados_brasil.head(3)
```

```python
"""Estamos criando  uma figura (fig) e um eixo (ax) usando plt.subplots(). 
Em seguida, o gráfico de linha é traçado com ax.plot(), e os rótulos dos eixos e o título 
são adicionados com set_xlabel(), set_ylabel() e set_title(). Por fim, o gráfico é exibido com plt.show()."""


fig, ax = plt.subplots(figsize=(8, 4))  # Criando a figura (fig) e o eixo (ax) com tamanho definido
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])  # Plotando os dados: anos no eixo X e número de imigrantes no eixo Y
ax.set_title("Imigração do Brasil para o Canadá \n1980 a 2013")  # Adicionando título com quebra de linha (\n) para subtítulo
ax.xaxis.set_major_locator(plt.MultipleLocator(5))  # Definindo intervalo de 5 anos no eixo X
ax.set_xlabel("Ano")  # Adicionando rótulo ao eixo X
ax.set_ylabel("Número de imigrantes")  # Adicionando rótulo ao eixo Y
plt.show()  # Exibindo o gráfico
```

```python
"""Criando duas visualizações (um gráfico de linha e um boxplot) usando a função subplots com axes."""

fig, axs = plt.subplots(1, 2, figsize=(15, 5))  # Criando duas visualizações lado a lado usando axs ao invés de ax

# Primeiro gráfico de linha
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])  # Adicionando o gráfico de linha com os dados
axs[0].set_title("Imigração do Brasil para o Canadá\n1980 a 2013")  # Adicionando o título
axs[0].set_xlabel("Ano")  # Adicionando o rótulo do eixo x
axs[0].set_ylabel("Número de imigrantes")  # Adicionando o rótulo do eixo y
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))  # Definindo intervalo de 5 anos no eixo x
axs[0].grid()  # Adicionando grade para melhor visualização

# Segundo gráfico: boxplot
axs[1].boxplot(dados_brasil['imigrantes'])  # Adicionando o boxplot com os dados
axs[1].set_title("Boxplot - Imigração do Brasil para o Canadá\n1980 a 2013")  # Adicionando o título
axs[1].set_xlabel("Brasil")  # Adicionando o rótulo do eixo x (representa o país)
axs[1].set_ylabel("Número de imigrantes")  # Adicionando o rótulo do eixo y
axs[1].grid()  # Adicionando grade para melhor visualização

plt.show()  # Exibindo os gráficos
```
